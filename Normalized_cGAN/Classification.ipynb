```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os, random, cv2
%matplotlib inline
```

```python
fake = pd.read_csv('classification_result_fake.csv', header = None)
real = pd.read_csv('classification_result_real.csv', header = None)
original = pd.read_csv('classification_result_original.csv', header = None)
fake_c = pd.read_csv('classification_result_fake_c.csv', header = None)
```

```python
original.shape
```

```python
max(real.to_numpy()), min(real.to_numpy())
```

```python
plt.figure(figsize = (12,6))
_ = plt.hist(real, bins = 50, color = 'tab:blue', alpha = 0.5)
_ = plt.hist(fake, bins = 50, color = 'tab:orange',alpha = 0.5)
_ = plt.hist(original, bins = 50, color = 'tab:green',alpha = 0.5)
_ = plt.hist(fake_c, bins = 50, color = 'tab:red',alpha = 0.5)
plt.xlabel('Prediction', fontsize = 15)
plt.ylabel('Number of images', fontsize = 15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim((0,1800))
plt.xlim((0,1.1))
# plt.yscale('log')
```

## plot peak images

```python
plt.subplots(figsize=(4, 4))
gs1 = gridspec.GridSpec(4, 4)
gs1.update(wspace=0, hspace=0.05)
file = 'peak_fake'
for i in range(16):    
    img = cv2.imread(file+'/'+random.choice(os.listdir(file)))
    plt.subplot(gs1[i])
    plt.imshow(img, cmap = 'gray')  
    plt.axis('off') 
```
