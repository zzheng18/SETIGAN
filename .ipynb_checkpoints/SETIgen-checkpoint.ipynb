# Putting them in one .h5 file

```python
def plot_from_h5(h5_filename, data_name):
    hf = h5py.File(h5_filename, 'r')
    arr = np.array(hf.get(data_name))
    waterfall = np.flip(arr.reshape(128,128),1)
    plt.imshow(waterfall, aspect='auto', interpolation='none')
    plt.colorbar()
    plt.xlabel('Frequency (px)')
    plt.ylabel('Time (px)')
    plt.show()
```

# Putting them in seperate h5 files

```python
def plot_single_h5(filename):
    hf = h5py.File(filename, 'r')
    arr = np.array(hf.get('data'))
    waterfall = np.flip(arr.reshape(128,128),1)
    plt.imshow(waterfall, aspect='auto', interpolation='none')
    plt.colorbar()
    plt.xlabel('Frequency (px)')
    plt.ylabel('Time (px)')
    plt.show()
    hf.close()
```

```python
%%time
# generating and saving clean

for i in range(2500):
    
#     random_SNR = np.random.uniform(8, 20) 
    random_width = np.random.uniform(5, 30)
    random_drift = np.random.uniform(-0.3,0.3)
    if random_drift>0:
        random_drift = np.random.uniform(0.01,0.3)
        random_start = np.random.uniform(-10, 70)
    else:      
        random_drift = np.random.uniform(-0.3,-0.01)
        random_start = np.random.uniform(60, 140)

    frame = stg.Frame(fchans=128*u.pixel, # number of frequency samples
                  tchans=128*u.pixel, # number of time samples
                  df=2.7939677238464355*u.Hz, # frequency resolution
                  dt=18.253611008*u.s, # time resolution
                  fch1=6095.214842353016*u.MHz) # min/max frequency
#     noise = frame.add_noise(x_mean=10, x_std = 1, noise_type = np.random.choice(['normal', 'chi2']))
    signal = frame.add_signal(stg.constant_path(f_start=frame.get_frequency(index=random_start),
                                            drift_rate=random_drift*u.Hz/u.s),
                              stg.constant_t_profile(),
                              stg.gaussian_f_profile(width=random_width*u.Hz),
                              stg.constant_bp_profile(level=1))
    filename = 'seperate/clean_'+str(i+1)+'.h5'
    frame.save_h5(filename)

# generating and saving noisy

for i in range(2500):
    
    random_SNR = np.random.uniform(8, 20) 
    random_width = np.random.uniform(10, 40)
    random_drift = np.random.uniform(-0.3,0.3)
    if random_drift>0:
        random_drift = np.random.uniform(0.01,0.3)
        random_start = np.random.uniform(-10, 70)
    else:      
        random_drift = np.random.uniform(-0.3,-0.01)
        random_start = np.random.uniform(60, 140)
    
    frame = stg.Frame(fchans=128*u.pixel, # number of frequency samples
                  tchans=128*u.pixel, # number of time samples
                  df=2.7939677238464355*u.Hz, # frequency resolution
                  dt=18.253611008*u.s, # time resolution
                  fch1=6095.214842353016*u.MHz) # min/max frequency
    noise = frame.add_noise(x_mean=10, x_std = 1, noise_type = np.random.choice(['normal', 'chi2']))
    signal = frame.add_signal(stg.constant_path(f_start=frame.get_frequency(index=random_start),
                                            drift_rate=random_drift*u.Hz/u.s),
                              stg.constant_t_profile(level=frame.get_intensity(snr=random_SNR)),
                              stg.gaussian_f_profile(width=random_width*u.Hz),
                              stg.constant_bp_profile(level=1))
    filename = 'seperate/noisy_'+str(i+1)+'.h5'
    frame.save_h5(filename)
```

```python
%%time
# plotting clean
for i in range(2500):
    filename = 'seperate/clean_'+str(i+1)+'.h5'
    plot_single_h5(filename)
    
# plotting noisy
for i in range(2500):
    filename = 'seperate/noisy_'+str(i+1)+'.h5'
    plot_single_h5(filename)
```
